```python
#include <stdio.h>

int interpolationSearch(int arr[], int n, int key, int *probes)
{
    int low = 0, high = n - 1;
    *probes = 0;

    while (low <= high && key >= arr[low] && key <= arr[high])
    {
        (*probes)++;

        if (low == high)
        {
            if (arr[low] == key)
                return low;
            return -1;
        }

        int pos = low + (((double)(high - low) /
                         (arr[high] - arr[low])) *
                         (key - arr[low]));

        if (arr[pos] == key)
            return pos;

        if (arr[pos] < key)
            low = pos + 1;
        else
            high = pos - 1;
    }

    return -1;
}

int binarySearch(int arr[], int n, int key, int *probes)
{
    int low = 0, high = n - 1;
    *probes = 0;

    while (low <= high)
    {
        (*probes)++;

        int mid = low + (high - low) / 2;

        if (arr[mid] == key)
            return mid;

        if (arr[mid] < key)
            low = mid + 1;
        else
            high = mid - 1;
    }

    return -1;
}

int main()
{
    int n, key;

    printf("Enter number of students: ");
    scanf("%d", &n);

    int arr[n];

    printf("Enter sorted roll numbers:\n");
    for (int i = 0; i < n; i++)
        scanf("%d", &arr[i]);

    printf("Enter roll number to search: ");
    scanf("%d", &key);

    int interpProbes, binaryProbes;

    int interpResult = interpolationSearch(arr, n, key, &interpProbes);
    int binaryResult = binarySearch(arr, n, key, &binaryProbes);

    printf("\nInterpolation Search:\n");
    if (interpResult != -1)
        printf("Roll number found at position %d\n", interpResult);
    else
        printf("Roll number not found\n");
    printf("Probes = %d\n", interpProbes);

    printf("\nBinary Search:\n");
    if (binaryResult != -1)
        printf("Roll number found at position %d\n", binaryResult);
    else
        printf("Roll number not found\n");
    printf("Probes = %d\n", binaryProbes);

    return 0;
}
```
